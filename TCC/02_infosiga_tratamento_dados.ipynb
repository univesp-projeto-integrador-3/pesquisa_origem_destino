```python
import pandas as pd
import warnings
warnings.simplefilter("ignore")
```

```python
# Base de Óbitos
# essa base não está sendo usada no trabalho, apenas aqui para análise, visto que a base fatais já constam os óbitos
url_file = 'http://painelderesultados.infosiga.sp.gov.br/bases/obitos_publico.csv'
df_obitos = pd.read_csv(url_file, encoding='latin1', sep=';')

# filtra apenas óbitos de 2022, que serão objeto do trabalho
mascara = df_obitos['Ano do Óbito'] == 2022
df_obitos_2022 = df_obitos[mascara]
df_obitos_2022.head()
```

```python
## Base de Acidentes Não Fatais
url_file = 'http://painelderesultados.infosiga.sp.gov.br/bases/acidentes_naofatais.csv'
df_nao_fatais = pd.read_csv(url_file, encoding='latin1', sep=';')
df_nao_fatais
```

```python
# filtra apenas os registros com ano de acidente 2022
mascara = df_nao_fatais['Ano do Acidente'] == 2022
df_nao_fatais_2022 = df_nao_fatais[mascara]
# Inclui a coluna com o identificador do tipo de acidente
df_nao_fatais_2022['Tipo Acidente'] = 'Não Fatal'
df_nao_fatais_2022.head()
```

```python
### Base de Acidentes Fatais
url_file = 'http://painelderesultados.infosiga.sp.gov.br/bases/acidentes_fatais.xlsx'
df_fatais = pd.read_excel(url_file, engine='openpyxl')
df_fatais.head()
```

```python
# filtra apenas os registros com ano de acidente 2022
mascara = df_fatais['Ano do Acidente'] == 2022
df_fatais_2022 = df_fatais[mascara]
df_fatais_2022.head()
```

```python
### Base de Acidentes Fatais Rodoviários
url_file = 'http://painelderesultados.infosiga.sp.gov.br/bases/acidentes_fatais_rodovias.xlsx'
df_fatais_rodovias = pd.read_excel(url_file, engine='openpyxl')
df_fatais_rodovias.head()
```

```python
# filtra apenas os registros com ano de acidente 2022
mascara = df_fatais_rodovias['Ano do Acidente'] == 2022
df_fatais_rodovias_2022 = df_fatais_rodovias[mascara]
df_fatais_rodovias_2022
```

```python
# Faz o Merge das bases de acidentes fatais e fatais rodovias
df_fatais_2022_todos = pd.concat([df_fatais_2022, df_fatais_rodovias_2022])
# Inclui a coluna com o identificador do tipo de acidente
df_fatais_2022_todos['Tipo Acidente'] = 'Fatal'
df_fatais_2022_todos.head()
```

```python
# junta os dataframes fatais e não fatais de 2022

# filtra apenas as colunas de interesse
df_fatais_2022_todos = df_fatais_2022_todos[['ID', 'Data do Acidente', 'Lat (GEO)', 'Long (GEO)', 'Tipo Acidente']].head()
df_fatais_2022_todos.head()
```

```python
# filtra apenas as colunas de interesse
df_nao_fatais_2022 = df_nao_fatais_2022[['ID', 'Data do Acidente', 'LAT_(GEO)', 'LONG_(GEO)', 'Tipo Acidente']]
df_nao_fatais_2022.head()
```

```python
# renomeia as colunas, para uniformizar os dois dataframes
print(df_fatais_2022_todos.columns)
a_renomear = {
    'ID': 'id',
    'Data do Acidente': 'data_acidente',
    'Lat (GEO)': 'latitude',
    'Long (GEO)': 'longitude',
    'Tipo Acidente': 'tipo_acidente',
}
df_fatais_2022_todos.rename(columns=a_renomear, inplace=True)
df_fatais_2022_todos.head()
```

```python
# renomeia as colunas, para uniformizar os dois dataframes
print(df_nao_fatais_2022.columns)
a_renomear = {
    'ID': 'id',
    'Data do Acidente': 'data_acidente',
    'LAT_(GEO)': 'latitude',
    'LONG_(GEO)': 'longitude',
    'Tipo Acidente': 'tipo_acidente',
}
df_nao_fatais_2022.rename(columns=a_renomear, inplace=True)
df_nao_fatais_2022.head()
```

```python
# Faz a junção dos dois dataframes (acidentes fatais e não fatais)
# os dataframes possuem apenas as colunas de interesse
df_acidentes_2022 = pd.concat([df_nao_fatais_2022, df_fatais_2022_todos])
df_acidentes_2022
```

```python
# para facilitar o reprocessamento, guarda o dataframe num arquivo csv
# para ser utilizado nos próximo passos
df_acidentes_2022.to_csv('df_acidentes_2022.csv')
print('Arquivo csv salvo com sucesso.')
```
