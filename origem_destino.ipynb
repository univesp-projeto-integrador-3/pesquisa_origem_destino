```python
import pandas as pd
import os
import ntpath
from simpledbf import Dbf5


def read_dbf(file_path):
    file_name = ntpath.basename(file_path)
    print(f'Lendo arquivo {file_name}')
    dbf = Dbf5(file_path)
    df = dbf.to_dataframe()
    print(df.columns)
    print(df.info())
    print(df.head())
    print('Salvando csv da base de dados dbf -> ', end="->")
    df.to_csv(file_name.replace('.dbf', '.csv'), index=False)
    print('ok')
    return df
```

```python
file_path = os.path.join(
    'Pesquisa Origem Destino 1977',
    'Banco de Dados',
    'od77_CD.dbf'
)
df_1977 = read_dbf(file_path)
```
