```python
import pandas as pd
```

```python
# Abre os dois arquivos, para poder fazer o JOIN
file_path_ibge = 'OD_2017_fator_de_comparação_IBGE.xlsx'
df_ibge = pd.read_excel(file_path_ibge, engine='openpyxl')
df_ibge.head()
```

```python
file_path_pesquisa = 'OD_2017.xlsx'
df_pesquisa = pd.read_excel(file_path_pesquisa, engine='openpyxl')
df_pesquisa.head()
```

```python
df = df_pesquisa.merge(df_ibge, left_on='ID_PESS - Identifica Pessoa', right_on='ID_PESS')
df.head()
```

```python
df['FE_PESS - Fator de Expansão da Pessoa'].sum()
```

```python
# cria as novas colunas aplicando o fator de expansão pela projeção DO IBGE
df['FE_PESS_2023'] = df['FE_PESS - Fator de Expansão da Pessoa'] * df['FATOR_2017_2023']
df['FE_PESS_2025'] = df['FE_PESS - Fator de Expansão da Pessoa'] * df['FATOR_2017_2025']
df['FE_PESS_2035'] = df['FE_PESS - Fator de Expansão da Pessoa'] * df['FATOR_2017_2035']
df['FE_PESS_2040'] = df['FE_PESS - Fator de Expansão da Pessoa'] * df['FATOR_2017_2040']
df['FE_PESS_2045'] = df['FE_PESS - Fator de Expansão da Pessoa'] * df['FATOR_2017_2045']
df['FE_PESS_2050'] = df['FE_PESS - Fator de Expansão da Pessoa'] * df['FATOR_2017_2050']
df['FE_PESS_2055'] = df['FE_PESS - Fator de Expansão da Pessoa'] * df['FATOR_2017_2055']
df['FE_PESS_2060'] = df['FE_PESS - Fator de Expansão da Pessoa'] * df['FATOR_2017_2060']
df.head()
```

```python
df.to_excel(f'OD_2017_E_IBGE.xlsx', index=False)
print('Arquivo salvo com sucesso')
```
